```python
import altair as alt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


chrs = pd.read_csv("chr18_19.tab", comment="#", sep="\t", header=None)
chrs=chrs.sort_values(by=[0,1])
chrs.columns = ["chr_dna", "Start_dna", "End_dna", "chain","chr_rna", "Start_rna", "End_rna", 'expr']
chrs

ch=input()
start=int(input())
end=int(input())

chrs.to_csv('out.csv')

chrs_chose=chrs[chrs["chr_dna"] == str(ch)]
chrs_chose.to_csv('out1.csv')

chrs_coord=chrs_chose[(chrs_chose["Start_dna"] >=start) & (chrs_chose["Start_dna"] <= end)]
chrs_coord.to_csv('out2.csv')

```

```python
import altair.vegalite.v1 as alt
import altair as alt
alt.renderers.enable('notebook')
alt.data_transformers.enable('csv')
```

```python
chart1=alt.Chart(chrs).mark_bar(color="blue").encode(
    x=alt.X('Start_dna:Q', bin=alt.Bin(maxbins=100)),
    y=alt.Y("sum(expr):Q"),
)
chart1
```
